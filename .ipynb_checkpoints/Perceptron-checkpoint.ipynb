```python
import numpy as np
import pandas as pd

data = pd.read_csv("./iris.csv",header=None)
data.head()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data = data[data[4]!=3]
data
```

```python
data[4] = data[4].map({1:1,2:-1})
data
```

```python
class Perceptron:
    def __init__(self,alpha,times):
        self.alpha = alpha
        self.times = times
        self.w_ = np.ones(1+X.shape[1])
        
    def step(self,z):
        """
        阶跃函数
        """
        return np.where(z>=0,1,-1) 
    
    def fit(self,X,y):
        """
        模型训练
        ————————
        X[样本数][特征数]
        y[样本数]
        """
        X = np.asarray(X)
        y = np.asarray(y)
        
        self.loss_=[]
        for i in range(self.times):
            loss = 0
            for x,target in zip(X,y):
                y1 = np.dot(x,self.w_[1:])+self.w_[0]
                loss = loss + (y1!=target)
                #权重更新公式：w(j) = w(j) + 学习率 * (真实值 - 预测值) * x(j)
                self.w_[0] += self.alpha * (target - y1)
                self.w_[1:] += self.alpha * (target - y1) * x
            self.loss_.append(loss)
        
    def predict(self,X):
        return self.step(np.dot(X,self.w_[1:]) + self.w_[0])
                
        
```

```python
t1 = data[data[4]==1]
t2 = data[data[4]==-1]
t1 = t1.sample(len(t1),random_state=0)
t2 = t2.sample(len(t2),random_state=0)
train_X = pd.concat([t1.iloc[:40,:-1],t2.iloc[:40,:-1]],axis=0)
train_y = pd.concat([t1.iloc[:40,-1],t2.iloc[:40,-1]],axis=0)
test_X = pd.concat([t1.iloc[40:,:-1],t2.iloc[40:,:-1]],axis=0)
test_y = pd.concat([t1.iloc[40:,-1],t2.iloc[40:,-1]],axis=0)
p = Perceptron(0.1,10)
p.fit(train_X,train_y)
result = p.predict(test_X)
display(result)
display(test_y.values)
display(p.w_)
display(p.loss_)
```
